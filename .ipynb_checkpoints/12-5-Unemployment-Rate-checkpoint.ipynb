```python
import pandas as pd

df = pd.read_excel('/Users/flatironschool/Downloads/Unemployment.xls', header=None).dropna()
```

```python
df.columns=df.iloc[0]
# df.columns
```

```python
column_cleaning = df[['State', 'Area_name', 'Unemployment_rate_2010']]

only_wanted_columns = column_cleaning[1:]

new = only_wanted_columns.reset_index()
```

```python
just_counties = []
for county in new['Area_name']: 
    just_counties.append(county[:-4])
    
new['Counties'] = just_counties

final = new.drop(columns=['index', 'Area_name'])

final['Counties'].to_csv('counties_list_2', encoding='utf-8')
```

```python
# changing state names to from abbreviations to full state names 

states_list = []

for i in final['State']: 
    if i in states: 
        st = states[i]
        states_list.append(st)
        
states_list
```

```python
# making new col named 'Full States' to store full state names
final['Full States'] = states_list
```

```python
# Deleting old column names 'States'
final_drop = final.drop(columns=['State'])
```

```python
# # Merging 'Counties' & 'Full States'
final_drop['CountyState'] = final_drop['Counties'].map(str) + " " + final_drop['Full States']
```

```python
# # Deleting old 'Counties' and 'Full States' cols
final_final = final_drop.drop(columns=['Counties', 'Full States'])
```
