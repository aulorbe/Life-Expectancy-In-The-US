## Import data with County and States together

```python
df_wage = pd.read_csv('csv_files/wage_data_state.csv')
df_smoke = pd.read_csv('csv_files/smoking_FINAL.csv')
df_life = pd.read_csv('csv_files/life_expectancy_data_county.csv')
df_obesity = pd.read_csv('csv_files/obesity_FINAL.csv')
df_food = pd.read_csv('csv_files/food_stamps_FINAL.csv')
df_unemployment = pd.read_csv('csv_files/unemployment_FINAL.csv')
```

```python
# Clean life by state

# Doing this here because it's in a weird formate
df_life['CountyState'] = df_life['County'].map(str) + " " +df_life['State']

df_life.head()
```

```python
# merging food to life and creating a final dataframe 

final_df = pd.merge(df_life, df_smoke.drop_duplicates(subset=['CountyState']), on='CountyState')
final_df.tail()
```

```python
# clean final
final_df.rename(columns={'County_x':'County', 'State_x':'State'}, inplace=True)

final_df = final_df.drop(columns=['County_y', 'State Ab', 'State_y'])
final_df['CountyState'].dtype
```

```python
#take out extra space in final_df county state
final_df['CountyState'] = final_df['CountyState'].apply(lambda x: x.replace("  ", " "))
df_food['CountyState'] = df_food['CountyState'].apply(lambda x: x.replace("  ", " "))
```

```python
final_df.tail()
```

```python
# merging food to final dataframe

something = pd.merge(df_food, final_df, on='CountyState', how='left').dropna()
something.tail()

```

```python
# merging obesity to final dataframe

final_w_o = pd.merge(df_obesity, something, on='CountyState', how='left').dropna()
final_w_o.tail()
```

```python
# merging unemployment to final dataframe

final_w_un = pd.merge(df_unemployment, final_w_o, on='CountyState', how='left').dropna()
final_w_un.tail()
```

```python
final_w_un['State'] = final_w_un['State'].apply(lambda x: x.strip())
```
