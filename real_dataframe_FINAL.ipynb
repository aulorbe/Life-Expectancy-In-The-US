```python
#import modules
import re
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Import data with County and States together

```python
df_wage = pd.read_csv('csv_files/wage_data_state.csv')
df_smoke = pd.read_csv('csv_files/smoking_FINAL.csv')
# df_life = pd.read_csv('csv_files/life_expectancy_data_county.csv')
df_obesity = pd.read_csv('csv_files/obesity_FINAL.csv')
df_food = pd.read_csv('csv_files/12_11_food_life_FINAL.csv') # now = food stamps AND life expect merged already
df_unemployment = pd.read_csv('csv_files/unemployment_FINAL.csv')
```

```python
# stripping double spaces in vals of CountyState col in df_smoke
df_smoke['CountyState'] = df_smoke['CountyState'].apply(lambda x: x.replace("  ", " "))
```

```python
# checking that each val in df_smoke is equal to each val in df_food
# df_smoke.iloc[0]['CountyState'] == df_food.iloc[0]['CountyState']
```

```python
# merging df_food with df_smoke on CountyState
df = df_food.merge(df_smoke, left_on='CountyState',right_on='CountyState')
```

```python
# merging main df with df_obesity
df = df.merge(df_obesity, left_on='CountyState',right_on='CountyState')
```

```python
# merging df with df_uemployment
df = df.merge(df_unemployment, left_on='CountyState',right_on='CountyState')
```

```python
# cleaning up spaces in df State col
df['State'] = df['State'].apply(lambda x: x.strip())
```

```python
# cleaning up spaces in df_wage State col
df_wage['State'] = df_wage['State'].apply(lambda x: x.strip())
```

```python
# function to get the minimum wages data into df
def get_the_wage(state):
    return df_wage[df_wage['State'] == state]['Min Wage'].values[0]

df['min_wage'] = df['State'].apply(lambda x: get_the_wage(x))
```

```python
df = df.drop(columns=['CountyState', 'State Ab'])
```
