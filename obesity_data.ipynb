```python
#import modules
import re
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/Users/flatironschool/Downloads/Obesity.xlsx')
```

```python
df.head()
```

```python
# get the State, County an and Obesity
df = df.iloc[:,[0,2,45]].dropna()
df.head()
```

```python
# renaming cols
df.columns = ['State', 'County', 'Percent Obese']
df.head()
```

```python
# dropping first row
dropped = df.drop(df.index[0]).reset_index(drop=True)

dropped
```

```python
# dropping PR rows
new = dropped[dropped.State != 'Puerto Rico']
new
```

```python
# Merging State & County in 'CountyState' col
new['CountyState'] = new['County'].map(str) + " " + new['State']
```

```python
# Dropping 'State' and 'County'
final = new.drop(columns=['State', 'County'])
final
```

```python
final.to_csv('csv_files/obesity_FINAL.csv', encoding='utf-8', index=False)

```
