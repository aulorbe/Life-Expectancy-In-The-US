```python
import pandas as pd

df_july = pd.read_excel('/Users/flatironschool/Downloads/SNAP_zip/JUL 2010.xls', header=None)
df_jan = pd.read_excel('/Users/flatironschool/Downloads/SNAP_zip/Jan 2010.xls', header=None)
```

```python
# CLEANING JANUARY TABLE
```

```python
jan = df_jan.iloc[4:].drop(columns=[2,4,6,7,8,9,10,11,12,13]).dropna()
jan.columns = ['County Info', 'SNAP Public Assistance Participation', 'SNAP Non-Public Assistance Participation', 'Total SNAP PA and Non-PA']
```

```python
cleaned_counties = []
for county in jan['County Info']:
    county = county[8:11] + county[15:]
    cleaned_counties.append(county)
```

```python
jan['new county info'] = cleaned_counties 
```

```python
states = []
for county in jan['new county info']: 
    states.append(county[:3].strip())
    
jan['States'] = states
```

```python
just_counties = []
for county in jan['new county info']: 
    just_counties.append(county[3:])
    
jan['Counties'] = just_counties
```

```python
jan = jan.reset_index(drop=True)
jan.drop(columns=['new county info', 'County Info'])
```

```python
# CLEANING JULY SNAP TABLE
```

```python
july = df_july.iloc[4:].drop(columns=[2,4,6,7,8,9,10,11,12,13]).dropna()
july.columns = ['County Info', 'SNAP Public Assistance Participation', 'SNAP Non-Public Assistance Participation', 'Total SNAP PA and Non-PA']

cleaned_counties = []
for county in july['County Info']:
    county = county[8:11] + county[15:]
    cleaned_counties.append(county)
    
july['new county info'] = cleaned_counties 
    
states = []
for county in july['new county info']: 
    states.append(county[:3].strip())
    
july['States'] = states

just_counties = []
for county in july['new county info']: 
    just_counties.append(county[3:])
    
july['Counties'] = just_counties

july = july.reset_index(drop=True)
july.drop(columns=['new county info', 'County Info'])
```

```python
# COMBINING JAN AND JULY TABLES
```
